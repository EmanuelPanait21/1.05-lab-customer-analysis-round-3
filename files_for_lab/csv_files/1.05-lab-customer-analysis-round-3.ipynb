```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mkt_data = pd.read_csv('marketing_customer_analysis.csv')
mkt_data
```

```python
mkt_data.describe()
```

```python
mkt_data.isnull().values.any()
```

```python
#Show a plot of the total number of responses.
sns.countplot(mkt_data['Response'])
plt.show()
```

```python
# Show a plot of the response rate by the sales channel.
sns.countplot('Sales Channel', hue = 'Response', data = mkt_data)
plt.show()
```

```python
# Show a plot of the response rate by the total claim amount.
sns.barplot(x="Response", y = "Total Claim Amount", data = mkt_data)
plt.show()
```

```python
#Show a plot of the response rate by income.
sns.barplot(y='Income', x='Response', data = mkt_data)
plt.show()
```
